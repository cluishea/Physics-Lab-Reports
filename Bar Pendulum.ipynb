```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Experiment Data

```python
data = {'t1L':[15.75,15.71,15.31,15.28,15.44,15.72,17.16,19.65,27.16],
       't2L':[15.66,15.67,15.42,15.41,15.45,15.82,16.88,19.71,27.22],
        'MeantL':[0,0,0,0,0,0,0,0,0],
        'TL':[0,0,0,0,0,0,0,0,0],
        'lL':[45,40,35,30,25,20,15,10,5],
        'l^2L':[0,0,0,0,0,0,0,0,0],
        'lT^2L':[0,0,0,0,0,0,0,0,0],
        't1R':[15.91,15.46,15.24,15.18,15.54,15.58,16.93,19.38,26.09],
        't2R':[15.79,15.66,15.25,15.21,15.16,15.59,17.12,19.54,26.07],
        'MeantR':[0,0,0,0,0,0,0,0,0],
        'TR':[0,0,0,0,0,0,0,0,0],
        'lR':[45,40,35,30,25,20,15,10,5],
        'l^2R':[0,0,0,0,0,0,0,0,0],
        'lT^2R':[0,0,0,0,0,0,0,0,0]        
       }
df = pd.DataFrame(data)
df
```

# Data Computation

```python
df['MeantL'] = (df['t1L']+df['t2L'])/2
df['TL']=df['MeantL']/10
df['l^2L']=df['lL']**2
df['lT^2L']=df['lL']*df['TL']**2

df['MeantR'] = (df['t1R']+df['t2R'])/2
df['TR']=df['MeantR']/10
df['l^2R']=df['lR']**2
df['lT^2R']=df['lR']*df['TR']**2
```

```python
df
```

# Plotting

```python
g = 978
k = 1/(12**0.5)*100
```

```python
# Plot of T vs l (Left taken as negative)

linspacexp = np.linspace(4,55,100)
linspaceyp = 2*3.14159*((linspacexp+k**2/linspacexp)/(g))**0.5

linspacexn = np.linspace(-55,-4,100)
linspaceyn = 2*3.14159*(((-linspacexn)+k**2/(-linspacexn))/(g))**0.5


plt.plot(np.append(linspacexp,linspacexn),np.append(linspaceyp,linspaceyn),c='g')


X = list(-np.array(df['lL']))+(list(df['lR']))
y = list(df['TL'])+(list(df['TR']))

plt.scatter(X,y,c='r')
plt.show()
```

```python
# Plotting lT^2 VS l^2

plt.scatter(df['l^2L'],df['lT^2L'],c='red')
plt.scatter(df['l^2R'],df['lT^2R'],c='orange')

X = np.append(df['l^2L'],df['l^2R'])
y = np.append(df['lT^2L'],df['lT^2R'])


from sklearn.linear_model import LinearRegression

X = X.reshape((-1,1))

model = LinearRegression()
model.fit(X,y)

b = model.predict(np.array([[0]]))
w = model.predict(np.array([[1]]))-model.predict(np.array([[0]]))
print(w,b)

linspacex = np.linspace(0,2000,100)
linspacey = w*linspacex+b

plt.plot(linspacex,linspacey,c='g')
             


plt.show()
```

```python
# Final Calculations

g = 4 * (3.14159)**2 / w
print(f"Acceleration due to gravity: {g}")
print(f"% error = {(g-978)/978*100}")
```
